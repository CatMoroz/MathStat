```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
Teta = 4
B = 0.95
etta = np.random.uniform(0,1,n)
sample = (1-etta)**(1/(1-Teta))
print(sample)
```

```python
S = 0
for i in sample:
  S += np.log(i)
Teta = (n/S)+1
print(Teta)
```

```python
a = -1.96*(((2**(1/(Teta-1)))*np.log(2))/((Teta-1)*(n**(1/2))))+2**(1/(Teta-1))
b = 1.96*(((2**(1/(Teta-1)))*np.log(2))/((Teta-1)*(n**(1/2))))+2**(1/(Teta-1))
L_confidence_interval_for_mediana = b-a
print(f"confidence_interval_for_mediana ({a},{b})")
```

```python
a = -1.96*((Teta-1)/(n**(1/2)))+Teta
b = 1.96*((Teta-1)/(n**(1/2)))+Teta
L_confidence_interval_for_param = b-a
print(f"confidence_interval_for_param ({a},{b})")
```

```python
N = 10000
Teta = 4
all_samples = []
all_tetas = []
for _ in range(N):
  etta = np.random.uniform(0,1,n)
  sample_ = (1-etta)**(1/(1-Teta))
  all_samples.append(sample_)
  S = 0
  for i in sample_:
    S += np.log(i)
    teta_ = (n/S)+1
  all_tetas.append(teta_)
all_samples = np.array(all_samples)
all_tetas = np.array(all_tetas)
print(all_samples)
print(all_samples.shape)
```

```python
K1 = int(N*(1-B)/2)
K2 = int(N*(1+B)/2)
parametric_bootstrap_interval = (all_tetas[K1], all_tetas[K2])
L_parametric_bootstrap_interval = all_tetas[K2]-all_tetas[K1]
parametric_bootstrap_interval
```

```python
N = 1000
Teta = 4
all_samples = []
all_tetas = []
for _ in range(N):
  sample_ = np.random.choice(sample, size=n, replace=True)
  all_samples.append(sample_)
  S = 0
  for i in sample_:
    S += np.log(i)
    teta_ = (n/S)+1
  all_tetas.append(teta_)
all_samples = np.array(all_samples)
all_tetas = np.array(all_tetas)
print(all_samples)
print(all_samples.shape)
print(all_tetas)
print(all_tetas.shape)
all_tetas.sort()
all_tetas

K1 = int(N*(1-B)/2)
K2 = int(N*(1+B)/2)

nonparametric_bootstrap_interval = (Teta + all_tetas[K1], Teta + all_tetas[K2])
L_nonparametric_bootstrap_interval = all_tetas[K2] - all_tetas[K1]
nonparametric_bootstrap_interval
```

```python
L_min_interval = min(L_confidence_interval_for_mediana, L_confidence_interval_for_param, L_parametric_bootstrap_interval)
L_max_interval = max(L_confidence_interval_for_mediana, L_confidence_interval_for_param, L_parametric_bootstrap_interval)
print(f"Длина доверительного интервала для медианы: {L_confidence_interval_for_mediana}", 
      f"Длина доверительного интервала для параметра: {L_confidence_interval_for_param}",
      f"Длина параметрического бутстраповского доверительного интервала: {L_parametric_bootstrap_interval}", 
      f"Длина непараметрического бутстраповского доверительного интервала: {L_nonparametric_bootstrap_interval}", 
      f"Макс. длина интервала: {L_max_interval}", f"Мин. длина интервала: {L_min_interval}", sep="\n")
```
